```python
from aixi_booster import *
import numpy as np
import pandas as pd
import random

import torch
import torchvision
import torchvision.datasets as datasets
```

```python
mnist_trainset = datasets.MNIST(root='./data', train=True, download=True)
```

```python
np.random.seed(0)
random.seed(0)
```

```python
#делаем датасет
#Суть датасета: есть рандомные массивы, надо найти максимум
X_lst = []
Y_lst = []
for i in range(30000):
    X_rec = np.ravel(np.array(mnist_trainset[i][0]))
    Y_rec = np.zeros(10)
    Y_rec[mnist_trainset[i][1]] = 1
    X_lst.append(X_rec)
    Y_lst.append(Y_rec)
X = np.array(X_lst)
Y = np.array(Y_lst,ndmin=2)
```

```python
border = 8000
X_train = X[:border,:]
X_test = X[border:,:]
Y_train = Y[:border,:]
Y_test = Y[border:,:]
```

```python
%%time
model = symbolic_regression_multyboosting(size_genom=900,tact_count=70, regularization=1.)
episodes=[]
for i in range(len(Y_train)-1):
    episodes.append([i,i+1])
model = symbolic_regression_multyboosting(size_genom=900,tact_count=30, regularization=1.)
model.fit(X_train,Y_train,forest_count=12,sample_part=0.7,episodes=episodes,popsize=30, maxiter=3, boosting_count=2, alpha_count=5, elitarism=2, mutation_p=0.02)
```

```python
#train
predicts = []
facts = []
i = 0
for episod in episodes[:border]:
    #мы учили систему выдавать последовательности длиной в 1, на более длинных система работать не обязана, 
    #так что тестим на последовательностях той же длины
    Y_pred = model.predict(X_train[episod[0]:episod[1],:],return_hist=True)
    [pred,p_array,m_pred] = Y_pred
    predicts.append(m_pred)
    facts.append(Y_train[i])
    i += 1
predicts = np.array(predicts)
facts = np.array(facts)
print('mae', np.mean(np.abs(predicts[:, 0] - facts)))
print('accuracy', np.mean(np.argmax(predicts[:, 0], axis=1) == np.argmax(facts, axis=1)))
print('BCE', torch.nn.BCELoss()(torch.clip(torch.tensor(predicts[:, 0]), 0, 1), torch.tensor(facts)).item())
```

```python
#test
predicts = []
facts = []
i = 0
for episod in episodes:
    #мы учили систему выдавать последовательности длиной в 1, на более длинных система работать не обязана, 
    #так что тестим на последовательностях той же длины
    if episod[1]>X_test.shape[0]:
        break
    Y_pred = model.predict(X_test[episod[0]:episod[1],:],return_hist=True)
    [pred,p_array,m_pred] = Y_pred
    predicts.append(m_pred)
    facts.append(Y_test[i])
    i += 1
predicts = np.array(predicts)
facts = np.array(facts)
print('mae', np.mean(np.abs(predicts[:, 0] - facts)))
print('accuracy', np.mean(np.argmax(predicts[:, 0], axis=1) == np.argmax(facts, axis=1)))
print('BCE', torch.nn.BCELoss()(torch.clip(torch.tensor(predicts[:, 0]), 0, 1), torch.tensor(facts)).item())
```
